```python
import time
import random
from tkinter import Tk
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
```

```python
tam = 10
class Examen():
    def inicio(self):
        option = Options()
        option.add_argument("--disable-infobars")
        option.add_argument("start-maximized")
        option.add_argument("--disable-extensions")
        option.add_experimental_option("prefs", {
            "profile.default_content_setting_values.notifications": 2
        })
        PATH = 'C:\driver-chorme/chromedriver.exe'
        self.driver = webdriver.Chrome(chrome_options=option, executable_path=PATH)

        self.datos = {'nombre': [], 'apellido': [], 'correo': [], 'estado': []}

    def fin(self):
        self.driver.quit()

    def get_nombres(self):
        self.driver.get("https://generadordenombres.online/")
        nom = []
        ape = []
        for i in range(0, tam):
            names = self.driver.find_element_by_id('nombreGenerado')
            name = str(names.text).split(sep=' ')
            nom.append(name[0])
            ape.append(name[1])
            self.driver.refresh()
        self.datos['nombre'] = nom
        self.datos['apellido'] = ape

    def get_correos(self):
        self.driver.get("https://10minutemail.net/")
        cor = []
        est = []
        for i in range(0, tam):
            self.driver.find_element_by_id("copy-button").click()
            cor.append(Tk().clipboard_get())
            est.append(bool(random.getrandbits(1)))
        self.datos['correo'] = cor
        self.datos['estado'] = est

    def to_correos(self):
        txt = ""
        for i in range(0, len(self.datos['correo'])):
            txt = txt + self.datos['correo'][i] + ","
        return txt

    def escribir(self, body):
        for i in range(0, len(self.datos['nombre'])):
            body.send_keys(self.datos["nombre"][i])
            time.sleep(1)
            body.send_keys(Keys.TAB)
            body.send_keys(self.datos["apellido"][i])
            time.sleep(1)
            body.send_keys(Keys.TAB)
            body.send_keys(self.datos["correo"][i])
            time.sleep(1)
            body.send_keys(Keys.TAB)
            body.send_keys(self.datos["estado"][i])
            time.sleep(1)
            body.send_keys(Keys.ENTER)
        time.sleep(1)

    def login(self, id, password):
        self.driver.get("https://www.facebook.com/")
        email = self.driver.find_element_by_id("email")
        email.send_keys(id)
        Password = self.driver.find_element_by_id("pass")
        Password.send_keys(password)
        self.driver.find_element_by_id("u_0_b").click()

    def post_content(self):
        time.sleep(1)
        self.driver.find_element_by_css_selector(".jm1wdb64 > .a8c37x1j").click()
        time.sleep(1)
        self.driver.find_element_by_css_selector(".notranslate")
        activepostarea = self.driver.switch_to_active_element()
        activepostarea.send_keys(Keys.CONTROL, 'v')
        time.sleep(2)
        activepostarea.send_keys("02/18/2021")
        e = self.driver.find_element_by_css_selector(
            ".k4urcfbm > .oajrlxb2 > .rq0escxv > .rq0escxv > .d2edcug0 > .a8c37x1j")
        e.click()
        time.sleep(5)

    def copiar_imagen(self):
        self.driver.get(
            "data:image/jpeg;base64,/9j/4AAQSkZJRgABAQAAAQABAAD/2wCEAAkGBxMTEhUTEhIVFhUXFhcVGBcVFxcXFxgVGBUWFxcWFRcYHSggGh0lHhcWITEiJSkrLi4uGB8zODMsNygvLisBCgoKDg0OGBAQFy0lHyUtLS0tLS0tLS0tLS0tLS0tLS0tLS4rKy8tLSsuLSstLS0tLS0tLS8tKy0tLSsrKy0rLf/AABEIAKgBLAMBEQACEQEDEQH/xAAbAAABBQEBAAAAAAAAAAAAAAAAAgMEBQYBB//EAD8QAAIBAgQDBgMGBAUDBQAAAAECEQADBBIhMQVBUQYTImFxgTKRoUJSscHR8BQjYuEHcoLC8SQzkhVTc6Oy/8QAGgEBAQADAQEAAAAAAAAAAAAAAAECAwQFBv/EADYRAAIBAgQDBQcDBAMBAAAAAAABAgMRBBIhMUFRYRMicYHwBTKRobHB0SPh8RRCUmIkcsJD/9oADAMBAAIRAxEAPwD3GgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgCgG775VJ6CaAhjQST6k1CkZsS7GLaf638K+w+JvoPOs1FLWT8kanOTdorze35f06guAB1usbh6NogPlbGnuZPnV7S3u6fX4hUk9Zu/0+H5u+o7exltDDXEU9GYAx6E1ioSeqTMpVIRdpSS8y0qGQUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUAUBxmA1NARb93MCoGhESfyFAMpaHPU9T+m1Qozexyg5VBd/upqR/mOy+5FZqm7XeiNcqqTstX09WXmN9xdf437tfu2j4v9Vw6/wDiB6mrmjHZX8fx+bmOWcvedlyX5/FvEes4K2ohbajnsCSepJ1J8zWLk3uzZGEYqyRa1ChQBQBQBQBQBQBQBQBQBQBQBQBQHCaAbOIWYzCTQDF3idtTBP4ULYBxFOU/SgsOJjFIJ1ABjXmfKhDi49DzoLDy3lOzD50AqaA7QBQBQBQBQBQBQBQBQBQEPEmXA5AT7k/2oCPdvmPAuc9AQB7sdvqfKiSb1ZJNpaK4z/Cu/wD3X0+5blV/1N8TfQeVZ5kvdXm/Vvr4mvJKXvvyX53fy8B25ctWU1KW0Hoo9AKkYzqSstWWU6dGF5NJfApLnaJ7py4OybnLvH8KD5xPpofI12rBxpq9eVui1Z5r9ozrPLhYZur0RW460VaMTxF0uQCVtBsqg7Dwx+A5V0U2mv0qCceb3OOtFxlbEYpqXKN7L4HoVeQfRBQBQBQBQBQBQBQBQBQBQDd2+q7mgIl3iI+yPnQtiJisc2WZ9hpziKFsU54subL3il8ubLP2ROvlsd+lUFglz7X9JP0j86cAeQ4ziWfFB7rX2sq7SVW5MZmQE5QPDnDAKusrUIeh9luIi7bYhHRQ3hFzcryYakwdxOutUpYce4kLFoSJMDQkiWbxEaAnn05VCIznZ9r98d/iHa2LhPd2QcuVB8JkHxkiGk9dAKpTSYJzPds2YgAzzy6zMen1oBjtBxo2QoS21265hLaEAwIliToqiRJPUVAHBu0t0v3V6y9polcxzo3krjQnyoQ0acSGxHyqiw7bxgPl60FiSHpYgqoAoAoAoAoBm5iI0Ak/QepoCLcTMZbpEDQf3qFGcbjrVlc1x1Qcp3P+UDU+1bKdKdR2grmmtiKdGOapKxSHjWIv6YSyQv8A713RfVRz+vpXZ/TUqOtaWvJevXM83+ur4jTDU9P8pbeXp+BVYRLLYj/qGfERobpMWVeR4QOa+cx5RqOirKrClemlFcv7rc/XxOShChUr2rSc2uL92/JerdBfEEY3A12Ua2SLVpJUKRqO7jVyRBkddRpFKai4ZY6p7t/flYVXNVVKWjjpGK28ud1+64F3huFG4ouOWtXG1cWyAC22cggwSAJHKuOWIyPLGzS2uelHCdos87xk90ufPzNRXIegFAFAFAFAFAFAFAFAIuXQu5igK+9xGfg2686FsYnttx+5au2rdtsrXHAkqW0kCYB132kUKR+F8ckqto4i8c5Dvct5EAABLagQDPhIBmN+dUGsxIBtnNqIM1GDzng2LSxjWtdzZWygKrdLE3GDsugmZ8TiRMiTVIeh4c+CY+xz8yP0oDD8Z7tm7g5bdxS91GKGWXOWBVxEGSc3mDI1qAuuxNzvLS5MndEhVyAjnDNqJ2/CqB/tSme/YBmC7NGkQqNoQd9SBUKRsdi172yBbY5gwFxVBy+HbNymI+VUFjwHB91bZ7jlnbQuxElF2k7R+lGCm7TOCrXExDq1zKtt7YzZQrAkLG4OpIGpC1AWHCrt53PeqmRSDaZSc7QCpLqdiSTp+NUFnj8ctlZY6k9J1JjbnQDGAx3eAsrlgDEgZR8vL50BcYXFmND7cqEsTLWPncH9aEsSbOIDbUA8DUB2gEudD6UBAsfCPMT7nWoUzmM4xeum4MOUt2rZKvffqN8o/seW016UMNTpqLqXcntFfc8WrjK1ZzVC0Yx3m/t6+BA4basNbe/lfEXl539omO8C6yg3O5Ecq21Z1FNUtIx6fS/P4GihSouk8Qr1J/7fW3LjxFYRnW4VZ2vG94XQGBkOkoAZTKCddoBG8RKkIuOaKy5dn+ef1LQqTU3Tm3LPuum2nK3w05j/ABbBCyn824O7jKoUfzbn9IGir0Jg6TtNShVdSXcjru+S6838jLE4eNCN6ku7sre9LpwS5bPyJeI4gXsobCm2SBaZiJezoIQjfWRB22POtFKlao1U1trbg+vI6q9dypRdFWvpfjHpzv8AzxRT4XFtZXJauXisknIFuDMTrJgwecec7k11zoRqyzSST+Bw0cVOjDJBtrwzfP7Ho9eOfQhQBQBQBQBQBQBQELEY0TC+k9PShbGS7Wm4+RFuFFJZnbScqx4QT6n5eVClbg8UisO5Z2gBmILNbZZhgWJIJAk6dKoO9teGOzW8TbPitSYgtOx2G+xHvUBS3+KkYlHN4d2RkIIuk6CVJC6D3jf50G/wwlf30FQFFjOzFo3u+C+ItJiOmk9RM6c5NAX2HQxB3AUfjVYMZxXEYDE3ltDE2u8DFQM2hLEBkzEZTMfDWx0KiWbK7GOZbGr4JwwWQFGyqdtpPh/M/KtfAp0m1iB4HVip0ZCpKnY/s1ZQlH3lYxhUhP3WmV+G4I6uWa4rTogClQJ3ZhmIJ9IG9QzLi7h1KG39krl6aRHKoDLcSS5aR1NnvUFx3UNogVvEEgAlvEWAAHSqC47PgtaS46ZHy5SvRh8cRoRMwRuNedAUnbbFQ9q0qZmbxEToQCNCNp3Pt50BScFYsz/wtwWGchWtujHKFMCNhJza6fZoQ9FwdjKiqWLQILHdiNz5UBmeN8avvdWxZy2mJOXvG7trgXfu4B/CgHuzfG8Qt9sNjB4tTaur8LgGCvsdJOvzFCG2sYkzrqKCxPBqA7QFeFysV5fEPQ8vY/iKhUY/AYDx3cIx1t3DeQHVbiMACGHMgQQeRPlXr1K2ka64qz6NevgfO0cP3p4VvWLzLk0+frRjy3UsYi2cQ+VzICW5bKCD4rraQu2gHIEmFmtc71KUuzjdLi/svXE3UstHER7WVpPZL/0/t0Wuh3i8l2tWosICGcp4blwnxBgw2TfWeR6VaKTgpT7zei5L9yYmUlUdOl3End83xv4fh8h7DWnxKC8BNwDuwWgK6zIZSAcpEkGBBIO+lYOSoPs5PTfqvE2qLxcVWitdtdn1W/nbTx0G+D4iyt97bMbjsoRmAPczJi1qTJJJ1bckgRMFiVUdOM7WXDn4/wAEwcqKqzpuWaWz/wAfDq/HfgM3rmJdjkumwq+EWlUeGOTba89NNo0rdGNGK70czet3xOepPEzl3JZEtMqW3r4cje15J9AFAFAFAFAFAcZoEmgMX/iD2jOHsMysV0MZQCxgbAHSZI321NCmXwfG3w5M4i7isrKt5e7DC2WTPoyDSIMgz7VSmyxFtMTaBBBVhmU76xHy1II6E1AeYcVxOIs4i5atrdZi0W1SBaQFVSGUCGEayT7UIek8KS73IFxwzAQDHymN6FKfEcDsqQr3go0Pds4UEgyCZ1IB2G1YSqwg0pSSvzZshQqTV4Rb8Ey6xIe3hrndGXFtmQ/FLZSV9azldrQtFQVWPae7dX8OJ5fb41jQpZsVdAWWIUh2EKWIZvhGg+GT/loqcaavUld8l6v9DvrVo152w9CKW2Z/DRXt56s9NtJduYQi4P5r2oYTBzFNiViDryjWtkXHOnwPJqKSTS3MM3DAWFoIO61HdScsZSEQWp0YNlOaB8JMmdc40qyr529L3vfSxvljsM8HkS79rJW1Uud+W56LcRxh2C63CuXfmFiZPmT8qwTjn6XOWWbJpvb5nnWJ7Om20q72SDp3wK//AHW5Wf7da9iOJUlZpPw/D1Pnp4Nwd03Hx/K0NtwIlbRd7jMokguwYhQo1DD4uZnzrxf1NqiWbjlul8Ht61Z71Cyhe7a66/Mbv8TZhmsmVIWAoDPqJLMp5QCNDuIrhxP9TGTy7bK2uvFu/A9DDdhUSu9euhJHEB3HevoNZIG4DFQVH9WkD+oVuVXLSz1PX8knRfa9nH1/A/w7HW7yBrZkQNNiumzLuKzhUU1dGudOUHaRX8c4AMQVbOUZYhgNRB3mazMB7hvA0tkMSXbm7RJ9AIA/uaoGeN464L1uwluVcFnc6KEAIKj+omABtBJ8qgKzEW8zlcUwIFxWs6qDIuLkKwATBZQRJ1HQ0ITOH2L7si3QCbROa6BCtMZYHWIJHImqQsuKcfw2H/7twA9B4j8htQpZ8H43bvpntOHXaRuPIg6/OhC2t3AdjQDOOXQMN11/08x++lSwM12oQ22tYtBJtEB4+1bb/k/+Vd2DampUJcdvH19DyvaUXSlDFRWsXr1T9fMYu8OYBnRlKODcN9yBCnUFzqWInQRGg22rbGunaMlqtMq+3I0TwslecWsr1zN8Ou9+nDwF4bFYc4QFR33dEKA4gKSSFLDlbE6TMAdRWtwqKu4vu35fbm2blVovCxmu/bnz68kut7LmRsLiLoYpduM/fnKVSRkkRNnoBIkdATy121KcPfgrZefHx9cvLRRr1bunVk3n000tw06L6XficUwPc2z3rIloaSg8dzTUW02UnXUnTppNWjV7WXcTcuuy8X6uY4jD9hC9SSjFct34Lg+vDlxL2ylm+i3O6z+ECWAzCPstmMyPzrgzVKbcc1j1slGslPLm03/k0FaTpCgCgCgCgOMwAk0BluN9qLSXFsyWc6hEUu0feIUGBQpUdsMH/FYUlPEQDoNyemux5a7UBScPxufDW8Q+dHs22a4FQTmSQxXN96COutUGh7GM5wqF4E+IAclOoH76UKWJw1tJYwo3JJgD3NEDmEx9t84tHMVE6bE8oNYt6tcUbeyajGUtmVVjhjEk3GYEn7MEnQeMyDrPLbT5cuHwsFTUqkIuctZ5knv/AGrouhqrS7WpJybsnaKTaSS2/LLbheFNtMpM89oAPOBy1135mt1KiqV4xfdvouS5X4rl8DJzlNLO7tK1+L8evAbPA8P3gud0A0ycvhVjEAuo0aPOs8qvcz7eeXLfT6eHInrz9f0rI1BlG8a9aEsLYaAdB9TqfxoBJoBNu2FEAADoNBrvS99wkloivxnBrT+IDI4OYMvXqRWam1oa5Uot3WjFHhgeyLV45+p2M6xHoDFaa9OFVZWtDdRnOk076mT4hwfEYVu8tO2WfjWJUaTnUzI/YjQDyKtKrhk5R1Xr4eO3M9WnXp11aS1Nhwi9cuWla6oUnpsw5Hyn1PrXpYWrKpSU5Kzfq/medWjGM2osj8T4plY2gGRzlCMQuRiROUHXKeXiXl711wUXxXVcUc8s2yXhyfEVZZMSgJlXWQy6yrHcEc1MfSk4OLsITUlcVg+H3Zm865R8KKN+jO51nyEeprEyDi/FUtKUkq5U5ABOvlHTfWoDF2ez1vD37Pf2nvpczq7sV7tHuSYNsDb+onSgNF2V4YmFvXLVli1pl70Loe7loyzuQdxPIVSGtTcEUBYkUBV3MOCHtMJUgj1RtvzHtSMnGSkt0YzhGcXCWz0MrgMM7o2Fcy2HcnIdrls7H1EyAdNQD1HrVJxjJVktJrfk/wB+PE8CjSnODw0nrTe3+UXt8OHDZdVIwpW1ft9+4tFlKLakNIbSLkDKqzAAk6jfedNXv024K9nv+OPi/wBjooJUq0Y1JKLa0jvfx4LovnumjH5u8ZMN/Jyt47kTdZtDAza5dRuYPpvlTScE6rvdaLZL9/W5rrTlGq44dZbPV2u2+WvDXz4aDwRsSBfVczL/AC8pgLIP/ctM3wgg67kERuKwUlQ/Sk7cb/Z/bgbZQeKSrwV3tZ/VPh14+asU2Newjst2/ezz4hh4yA/dJJGZhtMDQAcq64xqTinGmmv9t/2OGdSjTk4zrST4qG37v+OB6dXiH0wUAUAUAUBV43EZpA2Ee+8n8KFRirGG7i/iDbtTcuMHN1jp3REQCfulT4R94HnQo3wbjNoMoBLd+0MzIUztHhddACCoAnXQCqC34nbsWEe7eIW3IzHkSxAEx1JA96JN7EbSIXZfiatcNpLTIgUMjM05oPiEDQQGSCCQdelaaeIpTeWMteXrwejs+Ni6lpjeDpduZ7hZgPhUeEDrqNZ357GK6FNpWRLE7D2FQZUUKvQCBWDd9WUXQAN6A7QHE29z+NAdyzA6kD250B12kk9agEc/r+n78qoFGgEHf0/GgO0AAT6fj/agFigK/i/DFviNARPiygn4GUCd/tT7VMsbSutWrXF3mg76Rle3Pf8AJLw+GVJb7RAzNrrlHIcue31q3drEdrtpFU3aQB2U2LsKA2oh8mpLi2YJAgmASY5SCBhnV7GGaeVzyPKuJWds8H/LOJtkEFV1BjwH7rAE+Lw6gjlvEVmZ3vqNdm7JxFy8zl2t5QkOMsKxdhbYMCXADaNO8+gA1+AwSWly21gbk9T5k71SEgHzqXKTcPf5H5/rQhzHLEP00P8AlP6aGgMz2jBsXrWLXYHursfcOx9tfcLXfhH2tOVB+K8fX3PJx6dCtDFLb3ZeD9fQi3+FupY5VcsGZrzEBChGhcnYRuqgiByFbY14tJXt/qt79Py/mznnhKkZOVk29czeluvLwXyRI73Dthkck4gqRaB1XO+oVLkx4ROmfkeZOuqKqKtKC7vHwXT9vodE3R/p41Jd97X2u+CfT/t829Y+Cxd7W1duQLxyJ3Sx3RjQJ1UaAjcTM7kZ1aUE+0gr23vx9fP66qFeo4ulVlbNosqtl6Lovlz3aet8MuJ4VsCBzVkAOgkiTMeuulHiIS1c/qI4OpDuxh8Gja15p7YUAUAUBRdp+PphkAbMzOQioglnY/ZUfiToBQpUcK4z3hKvbNpxsrENKnYhhofTlVKd7QcOa8gFu4bZBmQAeREGdxrUBS4LhS2x3mJYILYCy+RV0KnMpAEk5EgnXSKA0VjEWMVbYK6XUIysAQwg9Ry96k43TjqvkwVHBey3dXO+uXWdxtBMc9ydTudNBrzrVToQgtEvJWS8vzcnU0tbinKA6aA5zoDpoDibUB1OvQH5nT8zQHCaA4n4/sUAMaABQABPp+P9qAWTQBHX5frQChQDWKsB0dCYDKykjcBgRI+dSSumjKE8slJcHcyVzhl7MpFq53wIlhcBslwwIuKIJCnNdYzqNV2bXmyzWy1+Xrc7KkaNWSlKp3VqlbXwe13oktduWqNVg8CEspaIBVFVddfhA6+ldKVkeeoqKtHYdt2FUkgatEnrGg+VUpVcQ41Dm3biQJZ2+FQdgANWYnlpp9YUzF3iGOL3suJBawguZRYIS6pn4SGJnSNvahDY9meKfxFhLpiSIYDbMN/351UC/stIynURQhAxWEFy29l9iCvt9lvUae4rOnUdOamuBrrUo1acqctmjL4NLl2x3D6nDsUuW9QXUfAwjVog+Hnp5V6dRwp1O0j/AH6p8ufx58DwqUalah2Mv/m7SjzXB9bcuPwJPDbIW6qXcttbiFBZYLNwcsyiQoGsaySSK015Xjmjq0/eXDz4nThKeWeSdkmn3Xa78uH88hnFuVuRhQFKvD3rnjYkHW0gbUL5DLPKN62Rjnh+q91olp5v0zTUn2dX/jx2esnq3/qr7L4dCU2Hu4v+bbxDWx8BTxQGXfKVIzA7g76+w1RqRw/cnBN89PwzolSqYt9pSquK2truvBr18Fs64D1goAoBvEXcqk/L1oDEdqcK7XLWJtWu9KFlZMwDBXiWSdMy+cSCdaFKcYaQxzLmYDvLigDxqsLpOhU5dDpEzvVKanhuIZ7as6wSBsZnzoDH/wCKOEuP/Cwua13hFwTC6xGc8hAYT6jnB2UWlJ33toYy2Kz/AA4CHGXDhlLW8rG9dEhCxjLaQaSJlgYnQ8jWde9knvy6erEjxseoVzmYCgA0B2oBBOo/fKqALUAkNQHc2nqfoP8Ak/KgEFqA7noDgagFLr6fj/agHQZ2oDoFAdoAnpQHcvWgOihCo7S8QuWbea2BsxLHcZRmIUQdSoeJB1A0NZQSckmYylYrMNxK9iGDJ4Va0MvUMT44LDUgqRqo0G2tc+J7s4xg+99jKnrO7WliE3Z6+b73CSVuW2Rh9oHMWUzMSA0ab5RW0Cr3A8YLifw7BU7trTG43hCk7rbE5mA2JIiOdAazgvDlw9pbS7KN+p5mqgXGH61SDeMWCH/0t6HY/P8AGowZrjP/AE+Kt4kaJc/lXegP2WPyH/j516GH/WoypPdar7r1zPIxf/GxMcQtpd2X2frkQsTwx87ZrZuXTq1w/Dl18WcwLYgbLqByNbY1oOFk7R5cfhx89Gc08NUVRtxvL/Lh430t5arkyff7k2bd1375iAk2jl75wCAGI10gzqNteQrRTdTtJQirLrrlXr1xOyrGj2Uas3mbt7umZ+XrmQ8HxTEIuU9wonwpkPgXkvhI9ee+9bKuGjKV4t+fE1YfGzhDLOK6JcFyN5XmnshQBQGZ7Y9oEwyZiCxmFRfiZzoABuaFRU8G4811il2xcsXBGjjwNImAw0J8p/OqUsXwKMSwlSYBKnKdDOsb7n51AVPFGu3L1uxZBFtSO9fUBQQYg7kxJHnlrBvM8p0QTpRVTR34evgaIqCIOo6HX51mc5y1aCiFUKOigAfIUsBYoAoANAJJoBDNQCHagEl6FFXmiB0A+e5+pNCDeehRJuVQGehBQu+dQEm1coB0fKgOxQHaAJoAoCPjcMLiFG2MH3Bkb1CNXQ1w7hlqwuW2sCS0kkkltySddYHyqybk7sq0VkTAKgHRQClFUg7baDVISnQMpB2IqApuIYLv7L2W3jLPRhqrfga2UKrpVFNcDRiqCr0pU3x+vAztm9cv4YI5MWibd9V0clfhJ5lYEEDWRXpSjClWcl/drF8Nd/26HjRnVxGGUJf26TXHTbxXPqSeD4QZ+7dVS266WmhWcrBDqkyCIPiME6aaTWnEVElni7y58PBv7HRhKLbdOUUoNaJ7vqlwtz0e2mgziLlpGKrav34JBcXWQZp1AClV08hvNbYqpJXcox6Wv9bmio6MJOKjKfXM1r5NL4I3leUe+FAIvXAoJOgFAZPivDQ921dgM1okgExuCMw8xPPrQyM3ibt2y5a5aYvfIzDPmt2lTRQDHmTAG5nqaoNPw9mKKWmSoOu4kbGgHcFg0t5sixmJY9ZJJj01MVioqOxsqVZTtmexJqms6KAKADUAkmqBtmoUaLUIILUKctasAdp19BqfpNUghruYk9SSfxNCiDc/flVAF/3++dANvd9vxoBC3Og+dC2JVi8dtTz06elYSkluzJQbJ2HvBtAdenP5VFJPZklBrdD9ZGAUAVAFUBQHQKA6BQDiioBaihBRrJEJNgaVAMYpIYNyPhPr9k/l70BmsbOGxi3BAt4gZGnYXR8JP0+bV6FP9fDuD3jqvDj68DyK18LjFUXu1NH/ANuHrxK6/g3NwlldsRuSNkgxmRvhRRBIJ1231Fb1ODhlTSh9fHi3zscjp1VVcmm6n08OCXK/z1RocNw9LiK7MCzAFmtNCM0QWEem9cDryi3FbLa+568cNGolOW73ts3zNFXOdYUBB4rdhQJ33oVFHa4hbZiiupI3WdR6VSkhkB8/WoCi49dxS3LS4ZCVzKXMKFVQTmzFiNI/KNakoyaumjnrdrdZCxt44m4FCEhiYYR4VA+JgTJEwNOc1mlpc3KRYCsTI7QBUAkmqBtmoBpmoUaDaj1FUDJeqDtpoDN5ZR6tp+GagGc2nr+FAINyNPc0A2WoUXGtCpHbzhFLHkPryFaKtTKtNzfSpZmR8XxApbCjdtW/KufLpqdSSbuQ8HijmDDQjWsbOLujOcVJWNfhMQLiK42I+ux+td0XdXPKnHLJoeqmJS8b46LUpaGe6ASQASEH3nj8PnXbhsG6lpTdo/XwPLx/tJULwprNP5Lq/wAfEhcC4vcS4bGJOrGVYkESdcsjQjXSNtq6MVhYSh2lHhujk9n4+pCp2OJer2fjwv8AT4Gnryj3xYFCigKAWBUILFCHYqohMQ6CgE37YZSp50BRcawH8Rh3tkePceVxdvnt6NW/DVuyqqXDj4HLjsN/UUJQ48PH1oUb458RhkzMQgBS/Bhs6x8c7KRrpzMV3KlCjWdt948rPl1R5Uq9TE4aN3aK0nzuufR/XQYtd9bGWwXt29wAbazO7ReIbU89jv1rKUKU23PV8d//ADoSnOvTio07qPD3f/TT+x6JXkH0AUBh+35c93D5LffKtw7eCdp5TtPnQphLF5P4y5dCIUZxZtd0xDm4jKtxngwACQIA1EHzqg23CMe7M6gSFciSfEogaf1CZ/YoUvZnf+xqNARbsBSSNzA9hsB8yferclh01CnQaA4aAQxoUZc0Aw7VQNK3iHqPxqgZPP8AfOhTmfwkTzB+hH50A09yN6Abe6N+XUaj3oWxwHpt+9RQpIzgCawlLKrmcIuTsRuI3AUAY859+tcbd9WdiVtEZnGcRbMc4J8xH4Vc/MzUXwF2OIaZUBk7kiIH61JyVtDOMb7m87OIRh0nnJ9ixiumiu4jzcQ71Gc7R3biYd2tGCIk8wswSOh8/Wu7BxhKslPb7nle0p1YYaUqTs19ONimwmIVil4W1VmXwC3bc3WdQFuHJbYIy5p+Mg16M4NXg5aJ63atbhq1dO3I8SnUjLLVUUm1pZPM2t9E0mr89Ss4/hoAIIY2/wCSzCNSu0/1dZ6joa6MNPVq1r6r1yOPHUtE075e634ffmX/AGU433o7q4f5ijQn7aj/AHD+/WvNx2E7N547P5Ht+yfaPbx7Ko+8vmvyaYCvNPbFAVSHRUAsCgFChGP2DyqkHaAg4hcrzybT/UNvmPwoCjxfZ2bjXLF97LP8YXVWPWJEGuynjLRUKkFJLa55tX2debqUqjg3vbZibfZHD73M9xjqXZzJPsayftGt/bZLkkYx9j4fed5Pm2zWVwnqAaAz+KVbgKuAQ0gg6ggzoQd6GRmT2WS0AMKiKqlmVTOUM28GZA12MjoKEJfDsMLCZmBLbsV11Jk+o13rVVqqlHM07dDZTpubsjuBxjtdEGVYkxIICgH3UjQedeXhMTWqV1reMrvg0lws1qnsmmd+IoU4UttVZcU78b81u00XkV7J5p0VAcNUHGNQo05qgYc0BHZqyAwzUKJd9ZB8/wC1ARrt3oR7g0MkiFjceLYlj7AH9awlUSNsKMpFBiu1qKdLc+ZbKfoDWHa8kbP6ey3G7Pbm2sZrTjrlKsPWDH0rJVOZi6PI0NriS3ba3LUlCCRIK689DHQ1orTu7I20o2TbKHiGKckkyoGgnLB9INaTcmM2rmbesWZRJ2Ew5LAKNWIA9ToKxWrsbb5U2z02zbCqFGygAegEV6SVlY8Nu7uF62GUq2zAqfQiDWUZOLTXAwnBTi4y2ehhMLiFspesXe8JDgBUuOmYSVaAJUn4TqNa+hlF1JQqQttxSfh1PjYSjQjUoVL3TtZNq/B6bcnqtSbjhNo94DaRrfhN0ojl0OYO6wIkELIBkLqa0wllldO9nra9kn1+eux01YucLSVk1ZXsm2uNrcdErJ3SGeynALzXEvv/AC7akMoI8dzpofhXzOp5RvXPjMdnTp09uL5nb7M9k9k1Vre9wXLx69DfgV5R7woCgFAUIKAoDsVSDlregH6Abv2gwg/8HkaAi9240Kg+YMfQ7UAsYduoHlE/WgJdAM4y5lRj0BoDz7H8TnDXMQRlILrbDkgZhc7tZjXVvpVMin4lxzE2CWGXKmQMQTDSuYlQR7a8xFCG04XjBcthm0n057UtcXJCYZQxcASRB/f72rTCjCE3OKs3a/kbJVZyiot6IdraYBFAE0AhqAZc0KR3NUEZ2qlEGY8jQEe7rpr6ChUabgnA1QB7gBc8jrl/U1DGUuR57/iXaK4i5/VlYehUfmDXNNd478PLuI81xM60NjI+AtNcuLaHxMwUe53/ADrJvS5geuWsOEthBoqqFA8gIrkbuzYUGPtS223tVuVI7awkkSY09j79awbNiRqey2BzXA52tjnzciB+Z+VbsPC8sxoxdTLDKuJsK7TzAoCh7QcJdj3+HJW8BBynKWB00PI/lXoYTExiuzq6x+h4/tHATm+2oaT2dtLr8nOC9mFQi7fPeXJkA6qp66/E3mduXWsMTjHV7sdI8jbgfZsaH6k3mnz5eH5NMoriPUHAtQh2KEFAUAtVqkHBa60A4qxQHaAKAKAKAKARet5lIoDA8d4PdJyFT3JfMwAn7WbwkEbkfU0KQ/8A0BMRdLEXBbBDFSxCsy7SvTfprrQodq7luxaCMPA7CQA2giFYwICg66kfDAma6MM8ssyMKmGeIi4dPnwJ/Cbr28OFDF7hKBBcMkBgoEgGIADGB0PM1lUyzqN8NdjnhGVGhFK7k7b89vktTQ1ynWFQHCKoG3oUYuUBHuVQR3NUo01ATezeFD3sxEhBm1+9sv5/Koyy0RsKhqMD/irgJS3eA2lCfqv+6tc0deFlvE8nxVgdK1nUX3ZPgAR/4i4sNEIp3AOhYjqdh5E9a1ylpZFSNFicTyrUVFdbtS1Ytm1Ey1hyzBFEsxgfr+dYpOTsiuSiszN1gMItpAi8tz1PMmvThFRVkePUm5ycmSKyMAoBSihR1VqAcC0ILAoQWts0A6toVQOUIFAFAFAFAFAFAFAFAFANXMOp3UfhQEDiHBUu23ttqjqVYHoR1qp2d0ZRk4tNES9wy4LitpkRSABM5jGsdAB9ayUkotcWanFualwS+YqKwNoUBw1ANtVKR3oCNcrIDBPTT6UKM3jvQqNF2SsRbZvvNHso/UmozGbLLG8St2h4216DU/KsHJLcRhKWxlO0PHFv2zaNoZTB1MkwZ0iIrROrfRI6qVHK73MlY4ZbDSF1BzKSZ08p6VqcmdBNxV8AedY3Mis7yTUMkgs3orEzNZ2Sws5rx/yr/uP4D5104aG8jjxdTaBpK6jhCgFgUKOItQDyJ0oQeWz1oS46qAVSCqAKAKAKAKAKAKAKAKAKAKAKAKAKAKAS9sHcA0BHuYFTtI+tC3I1zAsNoNQXIV5CNwR61Ska5QpGuCsgR2oUauCgRJ/9SuLbFpPCoGpG5OpOvLWuadR30OiNJbsrWuCTmPvWm5usQcReE+hrFsySIrX6xM0iNeeahlY5ZXrRmSC4ka1Cm87MLGFt+YY/N2rvpK0EeXXd6jLSthpCgIuJ4kic5PRY/wCKwlUjHc2xpTlsimTtNea/btqqIhuKpOrMQWAOpgDTy960dteSSOh4ZRg29zeAV1HAdoAoAoAoAoAoAoAoAoAoAoAoAoAoAoAoAoAoAoAoDhE70BFvcOttyj00+m1C3K7E8EP2GB8jp9aty3KrE4G4nxIY67j5iqW5BuuBrt+tYzkoq5nCOZ2K+9iuprhudqRAxF+sTJERrtDKw3NRmQGoUA1QEPEYw7VllI5Gh7KdtbSq2Hvypts2VgCwKliQCF1BE9K66ckopM4qtJuTaFcX/wARgHFvDYd3Y/buSiDzCjxN75azc0YRoN7jFvimKu63TI+4CFHy5+5rnlNs64Uox4Hb2JJ6jyI/c1pkzfEe4QM1+1/8if8A6FSHvIVfcfgerV6R4gUAUAUAUAUAUAUAUAUAUAUB/9k=")
      
        ele = self.driver.find_element_by_css_selector("img")
        ele.click()
        ele.click()
        activepostarea = self.driver.switch_to_active_element()
        activepostarea.send_keys(Keys.CONTROL, 'c')

    def ubicar_celdas(self, txt):
        seldas = self.driver.find_element_by_id('t-name-box')
        seldas.clear()
        seldas.send_keys(txt)
        time.sleep(1)
        seldas.send_keys(Keys.ENTER)
        time.sleep(1)

    def escribir_exel(self):
        self.driver.get(
            "https://docs.google.com/spreadsheets/d/1LsK2E-5Z65X5YL-f67Fg4OX1gTqb_nknV7Fn8VBskK8/edit#gid=0")
        #https://docs.google.com/spreadsheets/d/1LsK2E-5Z65X5YL-f67Fg4OX1gTqb_nknV7Fn8VBskK8/edit?usp=sharing
        #https://docs.google.com/spreadsheets/d/1LsK2E-5Z65X5YL-f67Fg4OX1gTqb_nknV7Fn8VBskK8/edit#gid=0
        self.ubicar_celdas("A2:D20")
        cel_del = self.driver.find_element_by_id('waffle-rich-text-editor')
        cel_del.send_keys(Keys.DELETE)
        time.sleep(1)
        self.ubicar_celdas("A2")
        body = self.driver.find_element_by_class_name('cell-input')
        self.escribir(body)

    def correo(self, correos, asunto="Flyer Elecciones"):
        self.driver.get(
            'https://login.live.com/login.srf?wa=wsignin1.0&rpsnv=13&ct=1612797357&rver=7.0.6737.0&wp=MBI_SSL'
            '&wreply=https%3a%2f%2foutlook.live.com%2fowa%2f%3fnlp%3d1%26RpsCsrfState%3db97197b5-ba7d-e675-321b'
            '-8f36488fb774&id=292841&aadredir=1&CBCXT=out&lw=1&fl=dob%2cflname%2cwld&cobrandid=90015')

        email = self.driver.find_element_by_name('loginfmt')
        email.send_keys('simulaciongut@hotmail.com')
        time.sleep(1)
        email.send_keys(Keys.ENTER)
        pasw = self.driver.find_element_by_name('passwd')
        pasw.send_keys('123UPS45')
        time.sleep(1)
        self.driver.find_element_by_id('idSIButton9').click()
        time.sleep(1)
        self.driver.find_element_by_id("id__5").click()
        time.sleep(2)
        para = self.driver.find_element_by_class_name("ms-BasePicker-input")
        para.send_keys(correos)
        time.sleep(1)
        para.send_keys(Keys.TAB)
        time.sleep(1)
        asu = self.driver.find_element_by_class_name("ms-TextField-field")
        asu.send_keys(asunto)
        time.sleep(1)
        asu.send_keys(Keys.TAB)
        mot = self.driver.find_element_by_class_name("_4utP_vaqQ3UQZH0GEBVQe")
        mot.send_keys(asunto)
        time.sleep(1)
        mot.send_keys(Keys.CONTROL, 'v')
        time.sleep(7)
        mot.send_keys(Keys.CONTROL, Keys.ENTER)
        time.sleep(2)
        self.driver.get("https://outlook.live.com/mail/0/sentitems")


if __name__ == '__main__':
    t = Examen()
    t.inicio()
    t.get_correos()
    t.get_nombres()
    print(t.datos)

    t.escribir_exel()
    t.copiar_imagen()
    t.login("fannys-cricri@hotmail.com", "Gutama093774481")
    t.post_content()
    t.correo(t.to_correos(), "Hola como estan esto es una prueba ")
    # t.fin()
```
